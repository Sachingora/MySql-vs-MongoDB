```python
import pandas as pd
import os

# Load datasets
tweets_df = pd.read_csv(r'C:\Users\Sachin Gora\OneDrive\Desktop\project\data\stocktweet.csv')
stock_prices_df = pd.read_csv(r'C:\Users\Sachin Gora\OneDrive\Desktop\project\data\AAPL.csv')

# Rename columns in stock_prices_df
stock_prices_df.rename(columns={"Date": "date", "Close": "close"}, inplace=True)

# Ensure date formats match
tweets_df['date'] = pd.to_datetime(tweets_df['date'], dayfirst=True)
stock_prices_df['date'] = pd.to_datetime(stock_prices_df['date'])

# Merge datasets on 'date'
merged_df = pd.merge(tweets_df, stock_prices_df, on='date', how='inner')

# Ensure the output directory exists
output_dir = r'C:\Users\Sachin Gora\OneDrive\Desktop\project\data'
os.makedirs(output_dir, exist_ok=True)

# Save the merged dataset
output_path = os.path.join(output_dir, 'merged_data.csv')
merged_df.to_csv(output_path, index=False)
print(f"Merged dataset saved to {output_path}")
```

```python
print(merged_df.isnull().sum())
```

```python
merged_df.fillna("Unknown", inplace=True)  # Example: Fill missing values
```

```python
import re

def clean_text(text):
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[^A-Za-z0-9\s$]', '', text)  # Remove special characters
    text = text.lower()  # Convert to lowercase
    return text

merged_df['tweet'] = merged_df['tweet'].apply(clean_text)
```

```python
merged_df = merged_df[['date', 'ticker', 'tweet', 'close']]
print(merged_df.head())
```

```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt

# Check columns in merged_df
print(merged_df.columns)

# Add sentiment analysis column if not present
if 'sentiment' not in merged_df.columns:
    merged_df['sentiment'] = merged_df['tweet'].apply(lambda x: TextBlob(x).sentiment.polarity)
    merged_df['sentiment'] = merged_df['sentiment'].apply(lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'neutral')

# Plot sentiment distribution
sentiment_counts = merged_df['sentiment'].value_counts()
sentiment_counts.plot(kind='bar')
plt.title('Tweet Sentiment Distribution')
plt.show()
```
